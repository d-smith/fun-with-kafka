## Starters

[This](https://docs.confluent.io/current/quickstart/ce-docker-quickstart.html) page provides a nice guide to kick the tires. Note the running this in docker requires a lot of memory. Fortunately you can run a large ec2 instance for something like 15 cents an hour.

To get started, launch an ec2 instances with sufficient memory - a general purpose instance type like a `t3a.xlarge` should work fine. The following assumes the use of the `Amazon Linux AMI 2018.03.0.20190826 x86_64 HVM gp2` AMI, which is the Amazon Linux 2 AMI.

To run the example from confluent, docker, docker compose, and git are required.

### Docker

```
 sudo yum install docker
sudo service docker start
docker ps
sudo usermod -a -G docker ec2-user
```

Note after the `usermode` you need to log out and log back in.

### Git

```
sudo yum install git
```

### Docker Compose

As per [this](https://docs.docker.com/compose/install/) page.

```
sudo curl -L "https://github.com/docker/compose/releases/download/1.25.0/docker-compose-$(uname -s)-$(uname -m)" -o /usr/local/bin/docker-compose
sudo chmod +x /usr/local/bin/docker-compose
```

## Confluent Example

Start up all the kit:

```
docker-compose up -d --build
```

Note the build option is to build the datagen source connector for use in the example.

Check out all the pieces:

```
docker ps
```

```
docker-compose down
```

Note that after a `docker-compose down` you can restart with a `docker-compose up -d`
